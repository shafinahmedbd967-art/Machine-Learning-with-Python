**ML Agorithm apply korar age amake decide korte hobe who's dependent and who's indepedent**

```python
from sklearn.model_selection import train_test_split
```

```python
x=dataset[["Hours of Study"]]  # indexing override er jonne eitay double bracket disi
x
```

```python

y=dataset["Exam Score"]
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
xtrain
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
# Ekhon Prothom kaj train kora
lr.fit(xtrain,ytrain)     # xtrain,yrain nsi karon ekta hocche proshno, arekta hocche uttor. mane oke amra sikhaitesi
# dataframe on dimensional
# series hocche multidimensional
```

```python
lr.predict(xtest)
```
