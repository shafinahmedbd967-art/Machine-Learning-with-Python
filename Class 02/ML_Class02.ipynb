```python
# abar dataset read korbo karon amake oi 3 ta column niye kaj korte hobe
# etake ekhn df naam dibo
df= pd.read_csv('titanic.csv',usecols=["Age","Cabin","Embarked"])
```

```python
df
```

**Mean diye replace**

Age e joto null value ase seitake mean, median, mode 3 ta diye replace korbo. erpr main dataset er sathe compare korbo graphicsl plot diye . dekhbo je kar poriborton sobcheye kom. jeitar porborton sobcheye kom seita diye e main dataset ke replace korbo

```python
# Mean diye replace
mean= df["Age"].mean()
```

```python
# Creating a Function
# diu= parameter ...eikhetre jehetu ei function ta dataset ke chine na sehetu dataset er variable diu
# Age ke pass korar jonne variable namr parameter use kortesi
# ei function er kaj Age_mean toiri kore oikhan er null value ke mean diye replace kora

def input_nan(diu,variable,mean):
  diu[variable+"mean"]=diu[variable].fillna(mean)
```

```python
# function call
input_nan(df,"Age",mean)
```

```python
df
```

```python
df.isnull().sum()
```

**Mode diye replace**

```python
# Mode diye replace so mode er man ber korbo
# [0] disi karon eita na dile index soho ashbe. mode diye replace hobe  na
mode= df["Age"].mode()[0]
mode
```

```python
def input_nan(diu,variable,mode):
  diu[variable+"mode"]=diu[variable].fillna(mode)
```

```python
# function call
input_nan(df,"Age",mode)
```

```python
df
```

```python
df.isnull().sum()
```

**Median diye replace**

```python
median= df["Age"].median()
median
```

```python
def input_nan(diu,variable,moedian):
  diu[variable+"median"]=diu[variable].fillna(median)
```

```python
# function call
input_nan(df,"Age",median)
```
