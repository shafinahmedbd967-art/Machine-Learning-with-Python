```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset=pd.read_excel('Online Retail.xlsx')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
dataset.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
```

```python
dataset['Description']=dataset['Description'].str.strip()
dataset.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
dataset['InvoiceNo']=dataset['InvoiceNo'].astype('str')
dataset=dataset[dataset['InvoiceNo'].str.contains('C')]
dataset.head()
```

```python
dataset['Country'].value_counts()
```

```python
mybasket=(dataset[dataset['Country']=='Germany'].groupby(['InvoiceNo','Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
mybasket
```

```python
def my_bas(x):
  if x<=0:
    return 0
  if x>=1:
    return 1
b_c=mybasket.applymap(my_bas)
mybasket=b_c
```

```python
mybasket.head()
```

```python
f_i=apriori(mybasket,min_support=0.07,use_colnames=True)
f_i
```

```python
rules=association_rules(f_i,metric='lift',min_threshold=1)
rules
```
